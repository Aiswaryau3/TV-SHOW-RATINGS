### ****2.7 Numeric Data Type Conversion and Mean Imputation**:**

* A list `numeric_cols` is defined to specify the columns intended to hold numerical data: **Total Seasons** and **Total Episodes**.
* The loop iterates over each column in this list to apply a uniform transformation.
* `pd.to_numeric()` is used to convert column values to numeric data types where possible.
* The parameter `errors='ignore'` prevents conversion failures from raising exceptions, preserving non-numeric values as they are.
* `df.loc[:, col]` ensures the assignment is applied safely to the entire column, avoiding chained assignment issues.
* This step improves data type consistency and ensures the dataset is suitable for accurate numerical analysis and modeling.



```python
numeric_cols = ['Total Seasons', 'Total Episodes']
for col in numeric_cols:
    df.loc[:, col] = pd.to_numeric(df[col], errors='ignore')
```

This code was performed to ensure that the columns **'Total Seasons'** and **'Total Episodes'** are treated as numeric values instead of text.  

- The list `numeric_cols` specifies which columns are expected to contain numeric data.  
- `pd.to_numeric(df[col], errors='ignore')` attempts to convert each of these columns to a numeric data type so that operations like aggregation, comparison, and plotting work correctly.  
- The `errors='ignore'` argument means that if a column cannot be safely converted (for example, it already has non‑numeric strings), it will be left unchanged instead of raising an error.

### ****2.8 Numeric Data Preprocessing and Mean Imputation Strategy**:**

* A loop is used to systematically preprocess the **Average Runtime** and **Rating** columns.
* `pd.to_numeric()` converts each column to a numeric data type to ensure analytical compatibility.
* The `errors='coerce'` parameter enforces data integrity by converting invalid or non-numeric entries into `NaN`.
* The column-wise mean is computed using `df[col].mean()`, which inherently excludes `NaN` values.
* Missing values in each column are imputed with the respective column mean to maintain dataset completeness.
* `df.loc[:, col]` guarantees safe and explicit assignment, avoiding chained assignment warnings.
* This approach standardizes numeric columns and prepares the dataset for reliable statistical analysis and modeling.



```python
for col in ['Average Runtime', 'Rating']:
    df.loc[:, col] = pd.to_numeric(df[col], errors='coerce')  # convert invalid to NaN
    mean_val = df[col].mean()  # compute mean ignoring NaN
    df.loc[:, col].fillna(mean_val, inplace=True)  # replace NaN with mean safely
```

This code was performed to clean and standardize the numeric data in the **'Average Runtime'** and **'Rating'** columns so they can be used reliably in analysis and modeling.  

- `pd.to_numeric(df[col], errors='coerce')` converts the column to a numeric type and turns any invalid or non‑numeric values into `NaN`, which helps detect and handle bad or inconsistent entries.  
- `mean_val = df[col].mean()` calculates the average of the valid numeric values, automatically ignoring the `NaN` entries.  
- `df.loc[:, col].fillna(mean_val, inplace=True)` then replaces all `NaN` values with this mean, ensuring there are no missing values left in these columns and making the dataset more suitable for statistical analysis and machine learning algorithms.

##**2.9 Premiere Date Standardization and Missing Value Handling:**



* The **Premiere Date** column is converted to a datetime format using `pd.to_datetime()`.
* The parameter `errors='coerce'` ensures that invalid or improperly formatted dates are converted to `NaT`.
* The column is then cast to a string data type for uniform representation.
* Any `NaT` values resulting from failed date conversions are replaced with the label **"Unknown"**.
* This process standardizes date values while clearly handling missing or invalid entries, making the column suitable for reporting and analysis.



```python
df['Premiere Date'] = pd.to_datetime(df['Premiere Date'], errors='coerce')
df['Premiere Date'] = df['Premiere Date'].astype(str).replace('NaT', 'Unknown')
```

This code was performed to clean and standardize the **'Premiere Date'** column so that date information is usable and missing/invalid dates are clearly labeled.

- `df['Premiere Date'] = pd.to_datetime(df['Premiere Date'], errors='coerce')` converts the column to proper datetime objects, and any invalid or unparseable dates are turned into `NaT`, which marks them as missing.  
- `df['Premiere Date'] = df['Premiere Date'].astype(str).replace('NaT', 'Unknown')` then converts the dates back to strings and replaces the `NaT` entries with the label `'Unknown'`, making missing premiere dates explicit and easier to interpret in reports or visualizations.

# **2.12 Genre Column Parsing and Standardization:**

* The `ast` and `pandas` libraries are imported to enable data parsing and manipulation.
* The **Genres** column is processed to convert string representations of lists into actual Python list objects using `ast.literal_eval()`.
* A null check (`pd.notnull(x)`) ensures that missing values are handled safely during the conversion.
* Entries that are empty or null are initially represented as empty lists to maintain consistent data types.
* A subsequent transformation replaces any empty lists with a placeholder value **['Unknown']**.
* This process standardizes the **Genres** column, ensuring consistent structure for accurate analysis, filtering, and visualization.



```python
import ast
import pandas as pd

#Convert string representation of list to actual Python list (if needed)
df['Genres'] = df['Genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

#Replace empty lists with a placeholder
df['Genres'] = df['Genres'].apply(lambda x: x if len(x) > 0 else ['Unknown'])

```

#**8. CORRELATION MAP**:
Displays how strongly numeric columns are correlated with each other.Helps identify positive, negative, or no correlation between features.

- **Purpose**: Visualizes how numeric variables in a dataset are related to each other.  
- **Data Used**: Only numeric columns (integers and floats) are considered.  
- **Correlation Matrix**: Calculates values between -1 and +1 showing strength and direction of relationships.  
- **Heatmap**: Displays these correlations with colors (blue = negative, red = positive).  
- **Annotations**: Each cell shows the exact correlation value (formatted to 2 decimals).  
- **Title & Display**: Adds a clear title and shows the heatmap for interpretation.  
- In short: It’s a quick way to spot strong positive or negative correlations among numeric features.



```python
#8:Correlation Heatmap (numeric vs numeric)
plt.figure(figsize=(10,8))
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

###What the graph shows:

* Shows relationships between *Average Runtime*, *Rating*, *Total Seasons*, and *Total Episodes*.
* Most features have **weak correlations** (close to 0).
* *Total Seasons* and *Total Episodes* show a **moderate positive correlation** (0.53).
* Indicates that more seasons usually mean more episodes, while other features are largely independent.

# **10.PIVOT HEATMAP**

###Pivot Heatmap: Average Runtime by Type & Language

* A **pivot table** is created using `pivot_table()` to calculate the **mean Average Runtime**.
* `Type` is set as the **row index**, and `Language` is set as the **column index**.
* The aggregation function `aggfunc='mean'` computes the **average runtime** for each Type–Language combination.
* A figure size of **10×6** is defined to ensure better readability.
* `sns.heatmap()` is used to visualize the pivot table as a **heatmap**.
* `annot=True` displays the numeric values inside each cell.
* `fmt=".1f"` formats the values to **one decimal place**.
* The `YlGnBu` color map highlights **lower to higher average runtimes** using a color gradient.
* A clear and descriptive **title** is added to explain the visualization.
* `plt.show()` renders the heatmap for interpretation and comparison.


```python
#10:Pivot Heatmap: Average Runtime by Type & Language
pivot_table = df.pivot_table(values='Average Runtime', index='Type', columns='Language', aggfunc='mean')
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Average Runtime by Type & Language')
plt.show()
```

###What the graph shows:
Average Runtime by Type & Language (Heatmap)

* The heatmap displays the **average runtime (in minutes)** of TV shows across different **Types** and **Languages**.
* The **x-axis** represents the **Language** of the shows, while the **y-axis** represents the **Type** (e.g., Animation, Documentary, Scripted).
* Each cell shows the **mean runtime value** for a specific Type–Language combination.
* **Darker shades** indicate a **higher average runtime**, while lighter shades represent **shorter runtimes**.
* The **Scripted** category appears across many languages and generally has **longer runtimes** compared to other types.
* **Welsh Scripted shows** have the **highest average runtime** among all combinations.
* **Documentary, Reality, and Game Shows** in English tend to have **moderate to high runtimes**.
* Some Type–Language combinations are **blank**, indicating **no available data** for those categories.
* The color bar on the right helps interpret the **runtime scale**, making comparisons easier.
* Overall, the visualization highlights how **runtime varies significantly by both show type and language**, enabling quick comparative insights.
